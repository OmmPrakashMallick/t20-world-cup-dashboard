```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the correct sheet
df = pd.read_excel("ICC Mens T20 Worldcup.xlsx", sheet_name='ICC Mens T20 Worldcup', engine='openpyxl')
df.columns = df.columns.str.strip()

# Matches played by teams
matches_played = df['1st Team'].value_counts() + df['2nd Team'].value_counts()
df['Highest Score'] = pd.to_numeric(df['Highest Score'], errors='coerce')
df['Best Bowler Figure(Wickets Taken)'] = pd.to_numeric(df['Best Bowler Figure(Wickets Taken)'], errors='coerce')

# Top batsmen
top_batsmen = df.groupby('Top Scorer')['Highest Score'].max().sort_values(ascending=False).head(10)

# Top bowlers
top_bowlers = df.groupby('Best Bowler')['Best Bowler Figure(Wickets Taken)'].max().sort_values(ascending=False).head(10)

# Match results (winners)
match_results = df['Winners'].value_counts()

# Plotting
sns.barplot(x=match_results.index, y=match_results.values, palette='Set2')
plt.xticks(rotation=90)
plt.title("Match Wins by Team")
plt.show()

# Top 10 Batsmen
sns.catplot(x=top_batsmen.values, y=top_batsmen.index, palette='Blues_d')
plt.title(" Top 10 Batsmen by Highest Score")
plt.xlabel("Highest Score")
plt.ylabel("Player")
plt.tight_layout()
plt.show()

# Top 10 Bowlers
sns.barplot(x=top_bowlers.values, y=top_bowlers.index, palette='Greens_d')
plt.title(" Top 10 Bowlers by Wickets Taken")
plt.xlabel("Wickets")
plt.ylabel("Bowler")
plt.tight_layout()
plt.show()

# Matches Played
matches_played = matches_played.sort_values(ascending=False)
sns.barplot(x=matches_played.index, y=matches_played.values, palette='Reds')
plt.xticks(rotation=90)
plt.title(" Total Matches Played by Each Team")
plt.xlabel("Team")
plt.ylabel("Matches Played")
plt.tight_layout()
plt.show()

# Toss decision counts
toss_decision = df['Toss Decision'].value_counts()
print(toss_decision)
```
